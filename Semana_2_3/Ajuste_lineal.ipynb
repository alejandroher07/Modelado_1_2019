```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import leastsq
```

```python
x=np.array([1.0,2.5,3.5,4.0,1.1,1.8,2.2,3.7])
y=np.array([6.008,15.722,27.130,33.772,5.257,9.549,11.098,28.828])
plt.plot(x,y,"or")
```

```python
funcLine=lambda params,x : params[0]*x+params[1]
```

```python
y1 = funcLine([10,-6],x)
plt.plot(x,y,"or")
plt.plot(x,y1,"-k")
```

```python
ErrorFunc=lambda tpl,x,y: funcLine(tpl,x)-y
p0 = [10,-6]
best_params,success = leastsq(ErrorFunc,p0,args=(x,y))
```

```python
print(best_params)
print(success)
```

```python
y2 = funcLine(best_params,x)
plt.plot(x,y,"or")
plt.plot(x,y2,":k")
```
