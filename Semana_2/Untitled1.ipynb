```python
x = np.arange(123)
y = np.arange(124)
yy,xx = np.meshgrid(x,y)
```

```python
print(xx)
plt.imshow(xx,interpolation=None)
plt.colorbar()
```

```python
print(yy)
```

```python
xc,yc = 49.5,49.5
R = np.sqrt((xx-xc)**2 + (yy-yc)**2)
```

```python
plt.imshow(R)
plt.colorbar()
```

```python
Psi = np.arctan((yy-yc)/(xx-xc))
```

```python
Psi = np.rad2deg(Psi)
Psi[:50,:] = Psi[:50,:]+90
Psi[50:,:] = Psi[50:,:]+270
plt.imshow(Psi)
plt.colorbar()
```
